**Renaming coloumns:**

```
import pandas as pd
import numpy as np
s1=pd.Series([9,11,1,3,3,5,6,2,9])
s2=pd.Series([1.1,3.5,4.5,8.6,2.9,4.6,7.7,5.6,4.8])
s3=pd.Series(['a','b','c','d','e','f','g','h','i'])
Data={'first':s1,'second':s2,'third':s3}
df=pd.DataFrame(Data)
df
            
```

```
df.rename(columns={"second":"column2","third":"column3"},inplace=True)
df
```

**Changing the row index:**

**Removing columns:**

```
import pandas as pd
import numpy as np
s1=pd.Series([9,11,1,3,3,5,6,2,9])
s2=pd.Series([1.1,3.5,4.5,8.6,2.9,4.6,7.7,5.6,4.8])
s3=pd.Series(['a','b','c','d','e','f','g','h','i'])
Data={'first':s1,'second':s2,'third':s3}
df=pd.DataFrame(Data)
df
```

```
#single column removing
df.drop(['first'],axis=1)
```

```
df.drop(columns=["first","second"])
```

```
df.drop(df.columns[[0,2]],axis=1)
```

**Removing rows:**

```
import pandas as pd
import numpy as np
s1=pd.Series([9,11,1,3,3,5,6,2,9])
s2=pd.Series([1.1,3.5,4.5,8.6,2.9,4.6,7.7,5.6,4.8])
s3=pd.Series(['a','b','c','d','e','f','g','h','i'])
Data={'first':s1,'second':s2,'third':s3}
df=pd.DataFrame(Data)
df
```

```
#removing single rows
df.drop(7)
```

```
#removing multiple rows
df.drop([7,8])
```
